```python
# dropping the lower case state column 
DFcsvMergEdit1=DFcsvMerg.drop(labels= 'state',axis=1)
```

```python
#changing the column names 
DFcsvMergEdit1.rename(columns=
              {'WhiteNonHispanic':'European',
              'BlackNonHispanic': 'African',
               'IndianNonHispanic':'Indigenous',
              'AsianNonHispanic':'Asian',
               'Hispanic':'Latinx',
               'costIndex':'Cost_Index'},inplace = True)
```

```python
DFcsvMergEdit2=DFcsvMergEdit1[['State', 'Cost_Index', 'Indigenous','European','Latinx','African','Asian']]
```
